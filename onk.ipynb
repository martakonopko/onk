```python
import matplotlib.pyplot as dupka
import pandas as pd
from matplotlib.ticker import (FormatStrFormatter) 

def create_plot(title, filename):
    fig, ax = dupka.subplots()
    fig.suptitle(title)
    data = pd.read_csv(filename, sep=';')
    for _, row in data.iterrows():
        variable = row[0]
        odds_ratio = float(row[1].replace(',', '.'))
        xmin, xmax = map(float, row[2].split(" to "))
        ax.hlines(y=variable,xmin=xmin, xmax=xmax, linewidth=2, color='#FF66CC')
        ax.plot([odds_ratio], [variable], marker='o', markersize=5, color="white",
                markeredgecolor="#FF66CC", markeredgewidth=2)
    ax.axvline(x=1, linestyle='--', linewidth=1, color="000")
    ax.set_xscale('log')
    ax.xaxis.set_major_formatter(FormatStrFormatter('%1.2f')) 
    return ax
    

create_plot("Przetoczenie KKCZ", "przetoczenie kkcz.csv")
create_plot("amoutacja", "amputacja.csv")
create_plot("reoperacja", "reoperacja.csv")
create_plot("TIMI major", "timi major.csv")
create_plot("TIMI minor", "timi minor.csv")
        
plt.show()
```
